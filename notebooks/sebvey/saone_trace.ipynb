```python
import pandas as pd
```

```python
path = '../../raw_data/saone_trace.csv'
```

```python
tracedf = pd.read_csv(path)[['X','Y','ele']]
```

```python
tracedf.columns = ['lon','lat','alt']
```

```python
def conc(x):
    return (x[0],x[1])
```

```python
tracedf['path'] = tracedf[['lon','lat']].apply(conc,axis=1)
```

```python
coord_list = list(tracedf['path'])
```

```python
coord_list
```

```python
final_dict = {
    'label':['Saône'],
    'color':['#7e8dbd'],
    'path':[coord_list]
}
```

```python
finaldf = pd.DataFrame.from_dict(final_dict)
```

```python
finaldf.to_pickle('saone_trace.pickle')
```
